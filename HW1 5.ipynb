```python
import numpy as np
import pandas as pd
from sklearn import linear_model as lm
import matplotlib.pyplot as plt
#from sklearn.linear_model import Ridge, RidgeCV

# open txt file
with open("D:/【CU】2022 Spring/5241/HW1/hw1_Q5_X.txt") as file_x:
    X = np.loadtxt(file_x, delimiter = " ")
with open("D:/【CU】2022 Spring/5241/HW1/hw1_Q5_Y.txt") as file_y:
    Y = np.loadtxt(file_y, delimiter = " ")
```

## a) Ridge regression

```python
ridge1 = lm.Ridge(alpha = 0.000001)
ridge1.fit(X, Y)
print(ridge1.coef_)
#print(ridge1.intercept_)
#type(ridge1.coef_)
#len(ridge1.coef_)
M1 = ridge1.coef_.reshape(128, 128)
plt.imshow(M1)
```

```python
ridge2 = lm.Ridge(alpha = 0.0001)
ridge2.fit(X, Y)
print(ridge2.coef_)
M2 = ridge2.coef_.reshape(128, 128)
plt.imshow(M2)
```

```python
ridge3 = lm.Ridge(alpha = 0.01)
ridge3.fit(X, Y)
print(ridge3.coef_)
M3 = ridge3.coef_.reshape(128, 128)
plt.imshow(M3)
```

```python
ridge4 = lm.Ridge(alpha = 0.1)
ridge4.fit(X, Y)
print(ridge4.coef_)
M4 = ridge4.coef_.reshape(128, 128)
plt.imshow(M4)
```

```python
ridge5 = lm.Ridge(alpha = 1)
ridge5.fit(X, Y)
print(ridge5.coef_)
M5 = ridge5.coef_.reshape(128, 128)
plt.imshow(M5)
```

### The best result is shown with $lambda$ = 1

## b) Lasso regression

```python
lasso1 = lm.Lasso(alpha=0.000001)
lasso1.fit(X, Y)
print(lasso1.coef_)
print(lasso1.intercept_)
N1 = lasso1.coef_.reshape(128, 128)
plt.imshow(N1)
```

```python
lasso2 = lm.Lasso(alpha=0.0001)
lasso2.fit(X, Y)
print(lasso2.coef_)
print(lasso2.intercept_)
N2 = lasso2.coef_.reshape(128, 128)
plt.imshow(N2)
```

## The best result is shown with $lambda$ = 0.0001

```python
lasso3 = lm.Lasso(alpha=0.01)
lasso3.fit(X, Y)
print(lasso3.coef_)
print(lasso3.intercept_)
N3 = lasso3.coef_.reshape(128, 128)
plt.imshow(N3)
```

```python
lasso4 = lm.Lasso(alpha=0.1)
lasso4.fit(X, Y)
print(lasso4.coef_)
print(lasso4.intercept_)
N4 = lasso4.coef_.reshape(128, 128)
plt.imshow(N4)
```

```python
lasso5 = lm.Lasso(alpha=1)
lasso5.fit(X, Y)
print(lasso5.coef_)
print(lasso5.intercept_)
N5 = lasso5.coef_.reshape(128, 128)
plt.imshow(N5)
```

## c)

We can see that the plot in ridge regression becomes clearer and clearer when lambda increases and gets to its best result when lambda is 1, while the plot in lasso regression only gets to its best when lambda is 0.0001 and then becomes less and less clear when lambda continously increase. Also, the best plot shown in lasso regression looks much clearer than ridge regression.
